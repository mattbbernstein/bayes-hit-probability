```python
import pandas as pd
import numpy as np
import pprint

pp = pprint.PrettyPrinter(indent=2)
```

```python
raw_data = pd.read_csv("data.csv")

# Remove sacrifices from dataset
data = raw_data.drop(columns=raw_data.columns[0])
data = data[~raw_data["events"].str.contains('sac_')]
data = data.copy()

# Create new column: True if a hit, False if not
hit_events = [
    'single',
    'double',
    'triple',
    'home_run'
]
data['is_hit'] = np.where(data['events'].isin(hit_events), True, False)
data.head()
```
